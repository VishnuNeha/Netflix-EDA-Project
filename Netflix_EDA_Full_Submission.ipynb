
# Netflix Titles Dataset - Exploratory Data Analysis

## Objective
Analyze Netflix content to uncover trends, patterns, distributions, and key insights using statistical summaries and visualizations.

### Tools Used
- Pandas
- NumPy
- Matplotlib
- Seaborn


```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['figure.figsize'] = (10,6)

df = pd.read_csv('data/netflix_titles.csv')
df.head()
```

```python

# Dataset Overview
print("Rows, Columns:", df.shape)
df.info()
```

```python

# Statistical Summary
df.describe(include='all').T
```

```python

# Missing Values Analysis
missing = pd.DataFrame({
    'Missing Values': df.isnull().sum(),
    'Percentage': round(df.isnull().mean()*100,2)
})
missing.sort_values('Missing Values', ascending=False)
```

```python

# Duplicate Records
print("Duplicate Rows:", df.duplicated().sum())
```

```python

# Content Type Distribution
sns.countplot(data=df, x='type')
plt.title('Movies vs TV Shows')
plt.show()
```

```python

# Rating Distribution
plt.figure(figsize=(12,6))
sns.countplot(data=df, y='rating', order=df['rating'].value_counts().index)
plt.title('Content Ratings Distribution')
plt.show()
```

```python

# Release Year Trend
year_counts = df['release_year'].value_counts().sort_index()
year_counts.plot()
plt.title('Content Released by Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

```python

# Top 10 Countries
countries = df['country'].dropna().str.split(', ')
countries = countries.explode()

countries.value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Content Producing Countries')
plt.show()
```

```python

# Top Genres
genres = df['listed_in'].str.split(', ')
genres = genres.explode()

genres.value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Genres')
plt.show()
```

```python

# Movies Duration Analysis
movies = df[df['type']=='Movie'].copy()
movies['duration_num'] = movies['duration'].str.extract('(\d+)').astype(float)

sns.histplot(movies['duration_num'], bins=30)
plt.title('Movie Duration Distribution')
plt.xlabel('Minutes')
plt.show()
```

```python

# Content Added to Netflix Over Time
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year

sns.countplot(data=df, x='year_added',
              order=sorted(df['year_added'].dropna().unique()))
plt.xticks(rotation=45)
plt.title('Content Added per Year')
plt.show()
```

```python

# Top Directors
top_directors = df['director'].value_counts().head(10)

top_directors.plot(kind='bar')
plt.title('Top 10 Directors on Netflix')
plt.show()
```

```python

# Correlation Analysis (Numerical Features)
numeric_df = movies[['release_year','duration_num']].dropna()

corr = numeric_df.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python

# Key Insights
insights = [
    "Movies represent around 70% of the Netflix catalog.",
    "United States is the largest content producer.",
    "India is among the top content-producing countries.",
    "TV-MA is the most common content rating.",
    "International Movies is the most frequent genre.",
    "Netflix catalog growth accelerated until 2018–2019."
]

for i, insight in enumerate(insights,1):
    print(f"{i}. {insight}")
```


# Conclusion

### Major Findings
- Movies dominate the Netflix library.
- Content is heavily concentrated in a few countries, especially the United States and India.
- Mature audience content (TV-MA) is most prevalent.
- International content is a major component of Netflix's strategy.
- The platform experienced rapid content expansion through 2019.
